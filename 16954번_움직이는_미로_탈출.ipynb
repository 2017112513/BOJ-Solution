```python
from collections import deque
arr = [list(input()) for i in range(8)]

block = deque()
next_block = deque()

for i in range(8):
    for j in range(8):
        if arr[i][j] =='#':
            block.append([i,j])
            if i+1!=8:  
                next_block.append([i+1,j])



def block_down():

    global block
    l = len(block)
    i= 0 
 
    while block:

        b_x,b_y = block.popleft()
        
        new_b_x = b_x + 1
        if new_b_x == 8:
            continue
        new_block.append([new_b_x,b_y])
 

        
    while next_block:
        next_block.popleft()
    
    for i,j in new_block:
        if i+1 != 8:
            next_block.append([i+1,j])



d = deque()
c_x, c_y = 7,0
d.append([c_x,c_y])

dx = [1,1,1,-1,-1,-1,0,0,0]
dy = [1,0,-1,1,0,-1,1,-1,0]


visited =[[[0]*8 for i in range(8)] for _ in range(9)]
# visited =[[0]*8 for i in range(8)]
ans_flag = False


d_flag= deque()
d_flag.append(0)
curr = 0

while d:
    
    num = d_flag.popleft()
    # print('curr',curr,'num  =',num)
    if curr != num and block:
        new_block = deque()
        # print(block,'한다')
        block_down()
        block = new_block
        curr+=1
    
    x,y= d.popleft()
    # print(x,y,'=----------------',curr,num)
    # print(block)
 
    if x==0 and y==7:
        ans_flag = True
        print(1)
        break

    for i in range(9):
        a = x+dx[i]
        b = y+dy[i]
        
        if (0<=a<8 and 0<=b<8) and i!=8 and visited[curr][a][b]==0:
                
            if [a,b] not in block and [a,b] not in next_block:
                # print(a,b)
                visited[curr][a][b] = 1
                d.append((a,b))
                d_flag.append(curr+1)
        elif (0<=a<8 and 0<=b<8) and i==8:
           
            if [a,b] not in block and [a,b] not in next_block:
                # print(a,b)
                d.append((a,b))
                d_flag.append(curr+1)

if not ans_flag:
    print(0)
    


#움직였을때? 그 쪽에 벽이 없으면 되잖아 
```
