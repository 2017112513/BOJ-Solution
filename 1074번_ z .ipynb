```python
n,r,c = map(int,input().split())
```

```python
cnt=0
def 사분면(n,r,c):
  global cnt
  if n==0:
    return 
  if r<2**(n-1):
    
    if c<2**(n-1):
      
      return 사분면(n-1,r,c)
    elif c>=2**(n-1):
      
      c= c-2**(n-1)
      
      cnt+=(2**(n-1))*(2**(n-1))
      return 사분면(n-1,r,c)
  elif r>=2**(n-1):
    
    if c<2**(n-1):
      r= r-2**(n-1)
      
      cnt+=(2**(n-1))*(2**(n-1))*2
      return 사분면(n-1,r,c)
    elif c>=2**(n-1):
      r= r-2**(n-1)
      c= c-2**(n-1)
      
      cnt+=(2**(n-1))*(2**(n-1))*3
      return 사분면(n-1,r,c)
print(cnt)
```
