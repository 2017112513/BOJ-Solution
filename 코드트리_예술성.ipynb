```python


from collections import deque
from collections import defaultdict

def get_score(array):
    for i in range(N):
        for j in range(N):
            dic[arr[i][j]].append([i,j])

    dx = [1,-1,0,0]
    dy = [0,0,1,-1]

    visited = [[0]*N for i in range(N)]

    total = defaultdict(list)
    group = 1

    for key,value in dic.items():

        sub_arr = dic[key]
        
        d = deque(sub_arr)

        while d:
            
            x,y = d.popleft()
            
            if visited[x][y]==1:
                continue
            
            flag = arr[x][y]

            sub_d = deque()
            sub_d.append([x,y])

            visited[x][y] = 1
            total[group].append([x,y])

            while sub_d:
                x,y = sub_d.popleft()
                
                for i in range(4):
                    a = x+dx[i]
                    b = y+dy[i]
                    if 0<=a<N and 0<=b<N and not visited[a][b]:
                        if arr[a][b] == flag:
                            visited[a][b] = 1
                            sub_d.append([a,b])
                            total[group].append([a,b])
                        else:
                            continue
                    else:
                        continue  
            group+=1

    dic2 = defaultdict(int)
    visited=[[0]*N for i in range(N)]

    group_list = [[0]*N for i in range(N)]

    cnt_list = [0]*100_000

    for key,value in total.items():
        
        for i,j in value:
            group_list[i][j] = key     

            cnt_list[key]+=1

    for key,value in total.items():
        d = deque(value)
        lis = []
        start = str(key)
        
        while d:
            
            x,y = d.popleft()

            for i in range(4):
                a = x+dx[i]
                b = y+dy[i]
                
                if 0<=a<N and 0<=b<N:
                    end = str(group_list[a][b])

                    if end==start:
                        continue

                    if end+','+start in dic2:
                        continue

                    dic2[start+','+end]+=1

    score = 0
 
    for key,value in dic2.items():
     
        n,m = key.split(',')
      
        n = int(n)
        m = int(m)
        i,j = total[n][0]
        p,q = total[m][0]
        point= (cnt_list[int(n)]+cnt_list[int(m)]) * arr[i][j] * arr[p][q] * value
        score += point
    
    return score


def roll(X,Y,n):
    x = X
    y = Y
    num = n
  
    next_x,next_y = y,x
    next_num = arr[next_x][next_y]
    arr[next_x][next_y] = num
    x,y,num = next_x,next_y,next_num

    next_x,next_y = N-1-X,x
    next_num = arr[next_x][next_y]
    arr[next_x][next_y] = num
    x,y,num = next_x,next_y,next_num
  
    next_x,next_y = y,x
    next_num = arr[next_x][next_y]
    arr[next_x][next_y] = num
    x,y,num = next_x,next_y,next_num

 
    next_x,next_y = abs(y-N+1),x
    next_num = arr[next_x][next_y]
    arr[next_x][next_y] = num
    x,y,num = next_x,next_y,next_num
   
def roll_array():
    new_arr = [[0]*N for i in range(N)]

    for i in range(N):
        if i<center:
            for p in range(center):
                for q in range(center):
                    new_arr[q][center-1-p]=arr[p][q]
                
                for q in range(center+1,N):
                  
                    new_arr[q-center-1][N-1-p]=arr[p][q]


        elif i>center:
            for p in range(center+1,N):
                for q in range(center):
                  
                    new_arr[q+center+1][N-1-p]=arr[p][q]

                for q in range(center+1,N):
                    new_arr[q][N-1-center-p]=arr[p][q]

    for i in range(N):
        for j in range(N):
            if i==center or j==center:
                new_arr[i][j] = arr[i][j]

    return new_arr

N = int(input())

dic = defaultdict(list)

arr = [list(map(int,input().split())) for i in range(N)]
answer = 0
center = N//2 
for i in range(4):
    answer += get_score(arr)

    for i in range(center):
        n = arr[i][center]
        roll(i,center,n)

    arr = roll_array()

print(answer)





    

        
    








# 5
# 1 2 2 3 3
# 2 2 2 3 3
# 2 2 1 3 1
# 2 2 1 1 1
# 2 2 1 1 1
```
