```python
N = int(input())
M=[]
for i in range(1,N):
    n=str(i)
    sum=0
    for j in range(len(n)):
        sum+=int(n[j])
    if i+sum==N:
        M.append(i)
if len(M)==0:
    print(0)
else:
    print(min(M))
```
