```python
from collections import deque
n,k = map(int,input().split())
lis = [i for i in range(1,n+1,1)]
lis = deque(lis)
print('<',end='')
while 1:
  for i in range(k-1):
    lis.append(lis[0])
    lis.popleft()
  if len(lis)==1:
    print(lis.popleft(),end='' )
  else:
    print(lis.popleft(),end=', ')
  if len(lis)==0:
    break
print('>')
```
