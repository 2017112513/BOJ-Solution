```python
from collections import deque
n= int(input())
lis = deque()
for i in range(n):
  d= int(input())
  if d == 0:
    if len(lis)==0:
      print(0)
    else:
      m = min(lis)
      print(m)
      lis.remove(m)
  else:
    lis.append(d)
```

```python
import heapq
n= int(input())
h=[]
for i in range(n):
  d = int(input())
  if d==0:
    try:
      print(heapq.heappop(h))
    except:
      print(0)
  else:
    heapq.heappush(h, d)
```
