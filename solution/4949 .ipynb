```python
from collections import deque
while 1:
    string = input()
    if string =='.':
        break
    check=True
    lis=deque()
    for i in string:
        if i == '(':
            lis.append('(')
        elif i== ')':
            if len(lis)!=0 and lis[-1]=='(':
                lis.pop()
            else:
                check=False
                break
        elif i=='[':
            lis.append('[')
        elif i==']':
            if len(lis)!=0 and lis[-1]=='[':
                lis.pop()
            else:
                check=False
                break
    if check and len(lis)==0:
        print('yes')
    else:
        print('no')
```
