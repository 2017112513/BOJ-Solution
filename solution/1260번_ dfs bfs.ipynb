```python
import sys
sys.setrecursionlimit(10**6)
n,m,v = map(int,input().split())
ans_d = [v]
def dfs(dic,v):
  if dic[v]:
    for i in sorted(dic[v]):
      if i not in ans_d:
        ans_d.append(i)
        dfs(dic,i)
      else:
        continue
ans_b = [v]
def bfs(dic,v,lis_key):
  if len(ans_b)==len(lis_key):
    return ans_b
  else:
    if dic[v]:
      for i in sorted(dic[v]):
        if i not in ans_b:
             ans_b.append(i)

    for j in sorted(dic[v]):
        bfs(dic,j,lis_key)
dic = {}
for i in range(m):
    a,b =map(int,input().split())
    try: 
        dic[a].append(b)
    except: 
        dic[a]=[]
        dic[a].append(b)
    try: 
        dic[b].append(a)
    except: 
        dic[b]=[]
        dic[b].append(a)
    lis_key = list(dic.keys())
dfs(dic,v)
bfs(dic,v,lis_key)
print(' '.join(list(map(str,ans_d))))
print(' '.join(list(map(str,ans_b))))
```
