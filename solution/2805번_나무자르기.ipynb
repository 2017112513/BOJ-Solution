```python
N,M=map(int,input().split())
namu=list(map(int,input().split()))

big = max(namu)
hubo=[]
sm=1
mid=(sm+big)//2
while sm<=big:
  sum=0
  for i in namu:
    if i>=mid:
      sum+=i%mid
  
  if sum>=M:
    hubo.append(mid)
    sm=mid+1
    mid=(sm+big)//2
    
  else:
    big=mid-1
    mid=(sm+big)//2
    
print(max(hubo))
```
