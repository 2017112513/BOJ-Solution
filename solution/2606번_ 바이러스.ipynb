```python
cnt = 0
def dfs(x):
  global cnt
  if visted[x]==0:
    visted[x]=1
    cnt+=1
    for i in lis[x]:
      dfs(i)
  else:
    return 

n = int(input())
m = int(input())
lis = [[] for i in range(n+1)]
visted=[0]*(n+1)
for i in range(m):
  a,b = map(int,input().split())
  lis[a].append(b)
  lis[b].append(a)

dfs(1)
print(cnt-1)
```
