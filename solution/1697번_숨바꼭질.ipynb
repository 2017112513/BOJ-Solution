```python
from collections import deque
def bfs(n):
  d = deque()
  d.append(n)
  while True:
    x=d.popleft()
    if x==k:
      print(lis[x])
      break
    for i in [x-1,x+1,2*x]:
      if 0<=i<=max and not lis[i]:
        lis[i]=lis[x]+1
        d.append(i)
n,k= map(int,input().split())
max=100000
lis =[0]*(max+1)
bfs(n)
```
