```python
T=int(input())
for i in range(T):
    n,k = map(int,input().split())
    아파트=[[] for i in range(n+1)]
    아파트[0]=[i for i in range(1,k+1)]
    i=0
    while i<n
        if n==0:
            print(k)
            break
        else:
            sum=0
            for j in range(k):
                sum=sum+아파트[i][j]
                아파트[i+1].append(sum)
        i+=1
    print(아파트[n][k-1])
```
