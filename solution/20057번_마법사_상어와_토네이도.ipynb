```python
N = int(input())
arr = [list(map(int, input().split())) for i in range(N)]
center_x, center_y = N // 2, N // 2


# 1 - left
# 2 - down
# 3 - right
# 4 - up


# new_arr = [[[] for i in range(N)] for j in range(N)]
# print(new_arr)
ans = 0

dx = [0, +1, 0, -1]
dy = [-1, 0, +1, 0]


direction_x = [
    [-2, -1, -1, -1, 0, +1, +1, +1, 2, 0],
    [-1, -1, 0, 0, 0, 0, +1, +1, +2, +1],
    [-2, -1, -1, -1, 0, +1, +1, +1, 2, 0],
    [-2, -1, -1, 0, 0, 0, 0, +1, +1, -1],
]
direction_y = [
    [0, -1, 0, +1, -2, -1, 0, 1, 0, -1],
    [-1, +1, -2, -1, +1, +2, -1, +1, 0, 0],
    [0, -1, 0, +1, +2, -1, 0, 1, 0, +1],
    [0, -1, +1, -2, -1, +1, +2, -1, +1, 0],
]
rate = [
    [0.02, 0.1, 0.07, 0.01, 0.05, 0.1, 0.07, 0.01, 0.02, 0.55],
    [0.01, 0.01, 0.02, 0.07, 0.07, 0.02, 0.1, 0.1, 0.05, 0.55],
    [0.02, 0.01, 0.07, 0.1, 0.05, 0.01, 0.07, 0.1, 0.02, 0.55],
    [0.05, 0.1, 0.1, 0.02, 0.07, 0.07, 0.02, 0.01, 0.01, 0.55],
]
dir = 0
next = 1
turn = 0
answer = 0
flag = False


while not flag:
    # print(center_x, center_y)
    for _ in range(next):
        # print("_ = ", _, "----------------------------------")
        # print("방향 : ", dir)
        sand = 0
        next_x, next_y = center_x + dx[dir], center_y + dy[dir]

        for i in range(10):
            # print("i=", i)
            # print(sand)
            a = next_x + direction_x[dir][i]
            b = next_y + direction_y[dir][i]

            if 0 <= a < N and 0 <= b < N:
                if i == 9:

                    arr[a][b] += arr[next_x][next_y] - sand
                    arr[next_x][next_y] = 0
                    # for j in arr:
                    #     print(j)

                    # print("-----------------------")
                    # print("answer", answer)
                else:
                    sd = int(arr[next_x][next_y] * rate[dir][i])
                    arr[a][b] += sd
                    sand += sd

            else:
                if i == 9:
                    answer += arr[next_x][next_y] - sand
                    arr[next_x][next_y] = 0
                    # for j in arr:
                    #     print(j)

                    # print("-----------------------")
                    # print("answer", answer)
                else:
                    answer += int(arr[next_x][next_y] * rate[dir][i])
                    sand += int(arr[next_x][next_y] * rate[dir][i])

        # if int(input()) == 1:
        #     pass
        if next_x == 0 and next_y == 0:

            flag = True
            break
        else:
            center_x, center_y = next_x, next_y

    dir += 1
  
    turn += 1

    if turn % 2 == 0:
        next += 1

    dir = dir % 4


print(answer)
```
