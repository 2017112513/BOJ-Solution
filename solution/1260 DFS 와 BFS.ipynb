```python
def dfs(v):
  print(v,end=' ')
  visted[v]=1
  for i in lis_ans[v]:
    if visted[i]==0:
      visted[i]=1
      dfs(i)
    else:
      continue

from collections import deque
def bfs(v):
  visted[v]=0
  a= deque([v])
  while a:
    p=a.popleft()
    print(p,end=' ')
    for i in lis_ans[p]:
      if visted[i]==1:
        visted[i]=0
        a.append(i)
      else:
        continue

     

n,m,v = map(int,input().split())
lis = [[] for _ in range(n+1)]
lis_ans=[]
visted=[0]*(n+1)
for _ in range(m):
  a,b= map(int,input().split())
  lis[a].append(b)
  lis[b].append(a)
for i in lis:
  lis_ans.append(sorted(i))

dfs(v)
print()
bfs(v)
```
