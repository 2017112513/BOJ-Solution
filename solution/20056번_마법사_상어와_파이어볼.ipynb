```python
# def direction(x, y, s, d):
#     if d == 0:
#         for _ in range(s):
#             new_x = x - 1
#             if new_x >= 0:
#                 x = new_x

#     elif d == 1:
#         for _ in range(s):
#             new_x = x - 1
#             new_y = y + 1
#             if new_x >= 0 and new_y < N:
#                 x = new_x
#                 y = new_y
#     elif d == 2:
#         for _ in range(s):
#             new_y = y + 1
#             if new_y < N:
#                 y = new_y
#     elif d == 3:
#         for _ in range(s):
#             new_x = x + 1
#             new_y = y + 1
#             if new_x < N and new_y < N:
#                 new_x = x
#                 new_y = y
#     elif d == 4:
#         for _ in range(s):
#             new_x = x + 1
#             if new_x < N:
#                 x = new_x
#     elif d == 5:
#         for _ in range(s):
#             new_x = x + 1
#             new_y = y - 1
#             if new_x < N and new_y >= 0:
#                 x = new_x
#                 y = new_y

#     elif d == 6:
#         for _ in range(s):
#             new_y = y - 1
#             if new_y >= 0:
#                 y = new_y
#     elif d == 7:
#         for _ in range(s):
#             new_x = x - 1
#             new_y = y - 1
#             if new_x >= 0 and new_y >= 0:
#                 x = new_x
#                 y = new_y

#     return x, y


def direction(x, y, s, d):
    new_x, new_y = x, y

    if d == 0:
        for _ in range(s):
            new_x = new_x - 1
            if new_x < 0:
                new_x += N

    elif d == 1:
        for _ in range(s):
            new_x = new_x - 1
            new_y = new_y + 1
            if new_x < 0:
                new_x += N
            if new_y >= N:
                new_y -= N

    elif d == 2:
        for _ in range(s):
            new_y = new_y + 1
            if new_y >= N:
                new_y -= N
    elif d == 3:
        for _ in range(s):
            new_x = new_x + 1
            new_y = new_y + 1
            if new_x >= N:
                new_x -= N
            if new_y >= N:
                new_y -= N
    elif d == 4:
        for _ in range(s):
            new_x = new_x + 1
            if new_x >= N:
                new_x -= N
    elif d == 5:
        for _ in range(s):
            new_x = new_x + 1
            new_y = new_y - 1
            if new_x >= N:
                new_x -= N
            if new_y < 0:
                new_y += N

    elif d == 6:
        for _ in range(s):
            new_y = new_y - 1
            if new_y < 0:
                new_y += N

    elif d == 7:
        for _ in range(s):
            new_x = new_x - 1
            new_y = new_y - 1
            if new_y < 0:
                new_y += N
            if new_x < 0:
                new_x += N
    return new_x, new_y


def solution():
    after_move = [[[] for _ in range(N)] for _ in range(N)]

    while ball:
        x, y, m, s, d = ball.pop()

        a, b = direction(x, y, s, d)
        # print(x, y, s, d, "->", a, b)
        after_move[a][b].append([m, s, d])
    # for q in after_move:
        # print(q)
    for i in range(N):
        for j in range(N):
            temp = after_move[i][j]
            if not temp:
                continue
            length = len(temp)

            if length >= 2:
                is_all = 0
                total_m = 0
                total_s = 0
                while temp:
                    m, s, d = temp.pop()

                    total_m += m
                    total_s += s

                    if d % 2 == 0:
                        is_all += 1
                    else:
                        is_all -= 1

                if abs(is_all) == length:
                    for k in range(0, 4):
                        new_m = total_m // 5

                        if new_m:
                            temp.append([new_m, total_s // length, k * 2])
                            ball.append([i, j, new_m, total_s // length, k * 2])

                else:
                    for k in range(0, 4):
                        new_m = total_m // 5
                        if new_m:
                            temp.append([new_m, total_s // length, k * 2 + 1])
                            ball.append([i, j, new_m, total_s // length, k * 2 + 1])
            else:
                m, s, d = temp.pop()
                ball.append([i, j, m, s, d])


N, M, K = map(int, input().split())

ball = []
for i in range(M):
    r, c, m, s, d = map(int, input().split())
    r, c = r - 1, c - 1
    ball.append([r, c, m, s, d])


for i in range(K):
    # print("원래", ball)

    solution()
    # print("이후", ball)
    # print("--------------------")


answer = 0
for i in ball:
    if i:
        answer += i[2]
print(answer)
```
