```python
n= int(input())
lis=[list(map(int,input().split())) for _ in range(n)]
one=0
zero=0
minus=0
def paper(x,y,n):
    global zero
    global one
    global minus
    start= lis[x][y]
    flag=True
    for i in range(x,x+n):
        for j in range(y,y+n):
            if lis[i][j]!=start:
                flag=False
                break
    if flag:
        if start ==1:
            one+=1
        elif start ==0:
            zero+=1
        else:
            minus+=1
    if not flag:
        for i in range(3):
          for j in range(3):
            paper(x+i*n//3,y+j*n//3,n//3)
paper(0,0,n)            
print(minus)
print(zero)
print(one)
```
