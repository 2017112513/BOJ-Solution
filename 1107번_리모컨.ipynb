```python
n = int(input())
m = int(input())
unfix = list(map(int,input().split()))
button = [i for i in range(0,10) if i not in unfix]
ans = abs(n-100)
for i in range(1000000+1):
  i_list= list(str(i))
  flag = False
  for j in i_list:
    if int(j) not in button:
      flag = True
      break
  if flag:
    continue
  else:
    ans = min(ans,abs(i-n)+len(str(i)))
    
print(ans)
  
```
