```python

from collections import deque

n,m = map(int,input().split())


lis = [list(input()) for i in range(n)]

def directions(i):
    if i==0:
        direction = 'down'
        return direction
    elif i==1:
        direction = 'up' 
        return direction
    elif i==2:
        direction = 'right'
        return direction
    else:
        direction = 'left'
        return direction


    

for i in range(n):
    for j in range(m):
        if lis[i][j] == 'R':
            red_x ,red_y = i,j

        elif lis[i][j] == 'B':
            blue_x ,blue_y = i,j  

dx = [1,-1,0,0]
dy = [0,0,1,-1]

d = deque()
d.append([red_x,red_y,blue_x,blue_y,0]) # count
visited = []
visited.append([red_x,red_y,blue_x,blue_y]) 
answer_flag = True
while d:

    RX,RY,BX,BY,cnt = d.popleft()

 
    if cnt>10:
        continue

    if lis[RX][RY] =='O':

      
        answer = cnt
        answer_flag = False
        break

    for i in range(4):
    
        rx,ry,bx,by = RX,RY,BX,BY
        
        direction = directions(i)
        blue_flag = True# blue_flag가 while문에서 false로 바뀌어 나온다면 파란공이 구멍에 빠진것이다 이럴 경우 큐에 추가하지않는다 
        red_flag = True  
        if direction == 'right':

            if rx-bx ==0: # R과 B가 같은 행에 존재한다면
               if ry<by : # B가 R보다 오른쪽에 존재한다면 
                    
                    while 1:

                        if lis[bx][by+1] == 'O':
                            by+=1
                            blue_flag = False
                            break

                        elif lis[bx][by+1]!='#':
                            by+=1
                        
                        else:
                            break

                    while 1:
                        if lis[rx][ry+1] == 'O':
                            ry+=1
                      
                            break
                        elif lis[rx][ry+1]!='#' and ry+1 != by :
                            ry+=1
                    
                        
                        else:
                            break
                    if [rx,ry,bx,by] not in visited and blue_flag:
                        d.append([rx,ry,bx,by,cnt+1])
               
                        visited.append([rx,ry,bx,by])
               else:
                                        
                    while 1:
                        if lis[rx][ry+1] == 'O':
                            ry+=1
                            red_flag = True
                            break

                        elif lis[rx][ry+1]!='#':
                            ry+=1
                        
                        else:
                            break

                    while 1:
                        if lis[bx][by+1] == 'O':
                            by+=1
                            blue_flag = False
                            break

                        elif lis[bx][by+1]!='#' and by+1 != ry:
                            by+=1
                        
                        elif by+1 == ry:
                            if not red_flag:
                                by+=1
                                blue_flag=False
                                break
                            else:
                                break   
                            
                        else:
                            break
                    
                    if [rx,ry,bx,by] not in visited and blue_flag:
                        d.append([rx,ry,bx,by,cnt+1])
                
                        visited.append([rx,ry,bx,by])
            else:# 두 공이 다른 행이 존재한다면 
            
                while 1: # 빨간공
                    if lis[rx][ry+1] == 'O':
                        ry+=1
                        red_flag = False
                        break

                    elif lis[rx][ry+1]!='#' :
                        ry+=1
                    
                    else:
                        break
                while 1: # 파란공
                    if lis[bx][by+1] == 'O':
                        by+=1
                        blue_flag = False
                        break

                    elif lis[bx][by+1]!='#':
                        by+=1
                    
                    else:
                        break
                
                if blue_flag:
                    if [rx,ry,bx,by] not in visited:
                        d.append([rx,ry,bx,by,cnt+1])
            
                        visited.append([rx,ry,bx,by])
        elif direction =='left':
           
            if rx-bx ==0: # R과 B가 같은 행에 존재한다면
                if ry>by : # B가 R보다 왼쪽에 존재한다면 
              
                    while 1:
                        if lis[bx][by-1] == 'O':
                            by-=1
                            blue_flag = False
                            break

                        if lis[bx][by-1]!='#':
                            by-=1
                        
                        else:
                            break

                    while 1:
                        if lis[rx][ry-1] == 'O':
                            ry-=1
                            break
                        
                        elif lis[rx][ry-1]!='#' and ry-1 != by :
                            ry-=1
                            
                        else:
                            break
                    if [rx,ry,bx,by] not in visited and blue_flag:
                        d.append([rx,ry,bx,by,cnt+1])
            
                        visited.append([rx,ry,bx,by])
                else: #B가 R보다 오른쪽에 있을 때 
               
                 
                    while 1:
         
                        if lis[rx][ry-1] == 'O':
                            red_flag=False
                            ry-=1
                            break

                        elif lis[rx][ry-1]!='#':
                            ry-=1
                        
                        else:
                            break
              
                
                    while 1:
                        if lis[bx][by-1] == 'O':
                            by-=1
                  
                            blue_flag = False
                            break 
                        
                        elif lis[bx][by-1]!='#' and by-1 != ry:
                            by-=1
                            
                        

                        elif by-1 == ry :
                            if not red_flag:
                                by-=1
                                blue_flag=False
                                break
                            else:
                                break   
                        else:
                            break
                    if [rx,ry,bx,by] not in visited and blue_flag:
                        d.append([rx,ry,bx,by,cnt+1])
                
                        visited.append([rx,ry,bx,by])

            else: # 두 공이 다른 행이 존재한다면 
               
                while 1: # 빨간공
                    if lis[rx][ry-1] == 'O':
                        ry-=1
                        red_flag = False
                        break

                    elif lis[rx][ry-1]!='#':
                        ry-=1
                    
                    else:
                        break
     
                while 1: # 파란공
                    if lis[bx][by-1] == 'O':
                        by-=1
                        blue_flag = False
                        break

                    elif lis[bx][by-1]!='#':
                        by-=1
                    
                    else:
                        break
                
                if blue_flag:
                    if [rx,ry,bx,by] not in visited:
                        d.append([rx,ry,bx,by,cnt+1])
         
                        visited.append([rx,ry,bx,by])
          
        elif direction =='up': #4 1 5 1
            if ry-by ==0: # R과 B가 같은 열에 존재한다면
            
                if rx>bx : # B가 R보다 위쪽에 존재한다면 
                    
                    while 1:
                        
                        if lis[bx-1][by] == 'O':
                            bx-=1
                            blue_flag = False
                            break

                        elif lis[bx-1][by]!='#':
                            bx-=1
                        
                        else:
                            break

                    while 1:
                        if lis[rx-1][ry] == 'O':
                            rx-=1
                            break

                        elif lis[rx-1][ry]!='#' and rx-1 != bx :
                            rx-=1
                        
                        else:
                            break
                    if [rx,ry,bx,by] not in visited and blue_flag:       
                        d.append([rx,ry,bx,by,cnt+1])
                      
                        visited.append([rx,ry,bx,by])
                else: # B가 R보다 아래쪽에 존재한다면
                    #4 1 5 1
                    while 1:
                        
                        if lis[rx-1][ry] == 'O':
                            rx-=1
                            red_flag= False
                            break

                        elif lis[rx-1][ry]!='#' :
                            rx-=1
                        
                        else:
                            break

                    while 1:
                  
                        if lis[bx-1][by] == 'O':
                  
                            bx-=1
                            blue_flag = False
                            break

                        elif lis[bx-1][by]!='#' and bx-1 != rx:
             
                            bx-=1
                        
                        elif bx-1 == rx :
        
                            if not red_flag:
                                bx-=1
                                blue_flag=False
                                break
                            else:
           
                                break   
                        else:
                            break
                    if [rx,ry,bx,by] not in visited and blue_flag:
                        d.append([rx,ry,bx,by,cnt+1])
               
                        visited.append([rx,ry,bx,by])

            else:# 두 공이 다른 열에 존재한다면 
                
                while 1: # 빨간공
                    if lis[rx-1][ry] == 'O':
                        rx-=1
                        red_flag = False
                        break

                    elif lis[rx-1][ry]!='#':
                        rx-=1
                    
                    else:
                        break
                while 1: # 파란공
                    if lis[bx-1][by] == 'O':
                        bx-=1
                        blue_flag = False
                        break

                    elif lis[bx-1][by]!='#':
                        bx-=1
                    
                    
                    else:
                        break
                
                if blue_flag:
                    if [rx,ry,bx,by] not in visited:
                        d.append([rx,ry,bx,by,cnt+1])
             
                        visited.append([rx,ry,bx,by])


        elif direction =='down':    
           
            if ry-by ==0: # R과 B가 같은 열에 존재한다면
                if rx<bx : # B가 R보다 아래쪽에 존재한다면 
                    
                    while 1:
                        if lis[bx+1][by] == 'O':
                            bx+=1
                            blue_flag = False
                            break

                        elif lis[bx+1][by]!='#':
                            bx+=1
                        
                        else:
                            break
                 
                    while 1:
                        if lis[rx+1][ry] == 'O':
                            rx+=1
                            break
                        
                        elif rx+1 != bx and lis[rx+1][ry] != '#':
                            rx+=1
                        
                        else:
                            break
                   
                
                    if [rx,ry,bx,by] not in visited and blue_flag:
                        d.append([rx,ry,bx,by,cnt+1])
               
                        visited.append([rx,ry,bx,by])
                else: # B가 R보다 아래쪽에 존재한다면
         
                    while 1:
                        if lis[rx+1][ry] == 'O':
                            rx+=1
                            red_flag = False
                            break

                        elif lis[rx+1][ry]!='#':
                            rx+=1
                        
                        else:
                            break
                    
                    while 1:
                        
                        if lis[bx+1][by] == 'O':
                            bx+=1
                            blue_flag = False
                            break
                        elif lis[bx+1][by]!='#' and bx+1 != rx:
                            bx+=1
                            
                        elif bx+1 == rx :
                            if not red_flag:
                                bx+=1
                                blue_flag=False
                                break
                            else:
                                break   
                        else:
                            break
                    if [rx,ry,bx,by] not in visited and blue_flag:
                        d.append([rx,ry,bx,by,cnt+1])
                    
                        visited.append([rx,ry,bx,by])
            else:# 두 공이 다른 열에 존재한다면 

                while 1: # 빨간공
                    if lis[rx+1][ry] == 'O':
                        rx+=1
                        red_flag = False
                        break

                    elif lis[rx+1][ry]!='#':
                        rx+=1
                    
                    else:
                        break
                while 1: # 파란공
                
                    if lis[bx+1][by] == 'O':
                        
                        bx+=1
                        blue_flag = False
                    
                        break

                    elif lis[bx+1][by]!='#':
          
                        bx+=1

                    else:
                 
                        break
                
                if blue_flag:
                    if [rx,ry,bx,by] not in visited:
                        d.append([rx,ry,bx,by,cnt+1])     
                     
                        visited.append([rx,ry,bx,by])     

        
                      
                      
if answer_flag:
    print(-1)
else:
    print(answer)
                      
                    




















    


```
