```python
N=int(input())
member=[(input().split()) for i in range(N)]
for i,j in enumerate(member):
  member[i]=[i,j]  
member_order= sorted(member,key=lambda x:(x[1][0],x[0]))
for i in range(N):
  print(' '.join(member_order[i][1]))
```
