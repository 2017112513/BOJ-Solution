```python
N=int(input())
덩치=[list(map(int,input().split())) for i in range(N)]
순위=[]
for i in 덩치:
  
  cnt=1
  덩치z= 덩치.copy()
  덩치z.remove(i)
  for h1,h2 in 덩치z:
    if (i[0]<h1) and (i[1]<h2):
      cnt+=1
    elif (i[0]<=h1) and (i[1]<h2):
      cnt+=1
  순위.append(cnt)
for i in 순위:
  print(i,end=' ')
```
