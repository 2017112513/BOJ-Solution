```python
n,m = map(int,input().split())
line = [int(input()) for i in range(n)]
upper,lower = max(line),1
while lower<=upper:
  cnt =0
  for i in line:
    mid=(lower+upper)//2
    cnt+=i//mid
  if cnt < m:
    upper=mid-1   
  elif cnt>=m:
    lower = mid+1
    ans = mid
```

```python
print(ans)
```
