```python
n = int(input())
for _ in range(n):
  N,M = map(int,input().split())
  a = list(map(int,input().split()))
  cnt=0
  target=a[M]
  a_index=[0 for i in range(6)]
  a_index[M]=target
  mx = max(a)
  while 1:
    mx = max(a)
    if a[0]==mx:
      cnt+=1
      if a_index[0]==target:
        print(cnt)
        break
      else:
        del a[0]
        del a_index[0]
    else:
      a.append(a[0])
      del a[0]
      a_index.append(a_index[0])
      del a_index[0]

```
