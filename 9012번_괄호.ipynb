```python
N=int(input())
for j in range(N):
    op=[]
    check=True
    string=input()
    for i in string:
        if i=='(':
            op.append(i)
        elif i==')':
            if len(op)!=0 and op[-1]=='(':
                op.pop()
            else:
                check=False
                break
    if len(op)==0 and check:
        print('yes')
    else:
        print('no')
```
