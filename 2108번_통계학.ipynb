```python
N = int(input())
lis = [int(input()) for i in range(N)]
mean=round(sum(lis)/len(lis))
print(mean)
lis_ordered=sorted(lis)
median_index = int(len(lis)/2)
median = lis_ordered[median_index]
print(median)
mode = {}
for i in lis:
  try: mode[i]+=1
  except: mode[i]=1
mode_sort=sorted(mode.items(),key=lambda x: [x[1],x[0]],reverse=True)
mode_value = mode_sort[0][1]
mode_list=[]
for i in mode_sort:
  if mode_value==i[1]:
    mode_list.append(i)
if len(mode_list)>1:
  print(sorted(mode_list)[1][0])
else:
  print(mode_list[0][0])
range1=max(lis)-min(lis)
print(range1)
```
