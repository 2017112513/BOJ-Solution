```python
sik = input()
ssum=0
if '+' in sik.split('-')[0]:
    n = list(map(int,sik.split('-')[0].split('+')))
    ssum+=sum(n)
else:
    ssum+=int(sik.split('-')[0])
for i in sik.split('-')[1:]:
  if '+' in i:
    n = list(map(int,i.split('+')))
    ssum-=sum(n)
  else:
    ssum-=int(i)
print(ssum)
```
